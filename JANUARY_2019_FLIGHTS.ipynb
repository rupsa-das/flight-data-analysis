#Advanced questions

```python
#Create a new column 'FLIGHT CODE' to uniquely identify a flight using 'OP_CARRIER' and 'OP_CARRIER_FL_NUM'.
df['OP_CARRIER_FL_NUM'] = df['OP_CARRIER_FL_NUM'].astype('str')
df['FLIGHT_CODE'] = df['OP_CARRIER'] + df['OP_CARRIER_FL_NUM']
df
```

```python
#Find the airport pair (origin → destination) with the highest average flight distance.
df['ROUTE'] = df['ORIGIN'] + '-' + df['DEST']
route_avg_distance = df.groupby('ROUTE')['DISTANCE'].mean()
route_avg_distance.sort_values(ascending = False)
route_avg_distance.head(1)
```

```python
'''
Create a FLIGHT_TYPE column based on distance:
<500 = Short-haul, 500–1500 = Medium-haul, >1500 = Long-haul
Then find which type of flights got delayed the most.
'''
flight_type = []

for dist in df['DISTANCE']:
  if dist < 500:
    flight_type.append('Short-haul')
  elif dist < 1500:
    flight_type.append('Medium-haul')
  else:
    flight_type.append('Long haul')

df['FLIGHT_TYPE'] = flight_type
df
types = df.groupby(by = 'FLIGHT_TYPE')['DEP_DEL15'].sum()
types.sort_values().tail(1)
```

```python
#Change the day of the week
days=[]

for day in df['DAY_OF_WEEK']:
  if day == 1:
    days.append('Monday')
  elif day == 2:
    days.append('Tuesday')
  elif day == 3:
    days.append('Wednesday')
  elif day == 4:
    days.append('Thursday')
  elif day == 5:
    days.append('Friday')
  elif day == 6:
    days.append('Saturday')
  elif day == 7:
    days.append('Sunday')

df['DAY'] = days
df
```

```python
df.drop('DAY_OF_WEEK',axis=1,inplace=True)
```

```python
df.rename(columns = {'DAY' :  'DAY_OF_WEEK'},inplace = True)
df
```

```python
#Plot number of flights cancelled per day of the week
import matplotlib.pyplot as plt
cancelled = df.groupby(by = 'DAY_OF_WEEK')['CANCELLED'].sum()
days = df['DAY_OF_WEEK'].unique()
plt.plot(cancelled.index,cancelled.values)
plt.xlabel("Days of the week")
plt.ylabel("Number of flights cancelled")
plt.show()
```

```python
#Create a column TIME_OF_DAY based on DEP_TIME_BLK (e.g., Morning, Afternoon, Evening). Which time of day has the most delays?
time_map = {
    '0001-0559': 'Night',
    '0600-1159': 'Morning',
    '1200-1759': 'Afternoon',
    '1800-2359': 'Evening'
}
df['TIME_OF_DAY'] = df['DEP_TIME_BLK'].map(time_map)

delayed = df[df['DEP_DEL15'] == 1]

delay_counts = delayed['TIME_OF_DAY'].value_counts()
most_delayed_period = delay_counts.idxmax()

print(f"Time of day with most delays: {most_delayed_period}")
```

```python
#Create a column DELAY_CATEGORY (None, Departure Only, Arrival Only, Both). What is the distribution of delay types?
df['ARR_DEL15'] = df['ARR_DEL15'].fillna(0)
df['DEP_DEL15'] = df['DEP_DEL15'].fillna(0)

delay_category = []

for dep,arr in zip(df['DEP_DEL15'],df['ARR_DEL15']):
  if dep==1 and arr==1:
    delay_category.append('Both')
  elif dep==1 and arr==0:
    delay_category.append('Departure only')
  elif dep==0 and arr==1:
    delay_category.append('Arrival only')
  elif dep==0 and arr==0:
    delay_category.append('None')

df['DELAY_CATEGORY'] = delay_category

import matplotlib.pyplot as plt
plt.plot(df['DELAY_CATEGORY'].value_counts())
plt.xlabel("Delay category")
plt.ylabel("Number of flights")
plt.show()
```

```python
#Create a column FLIGHT_DURATION_EST as ARR_TIME - DEP_TIME. What is the average estimated flight duration
df['FLIGHT_DURATION'] = df['ARR_TIME'] - df['DEP_TIME']
df['FLIGHT_DURATION'].mean()
```

```python
#Convert 'DEP_TIME' and 'ARR_TIME' columns into actual timestamps.
import numpy as np
df['DEP_TIME'] = df['DEP_TIME'].apply(lambda x: f"{int(x):04d}" if pd.notnull(x) else np.nan)
df['ARR_TIME'] = df['ARR_TIME'].apply(lambda x: f"{int(x):04d}" if pd.notnull(x) else np.nan)

from datetime import datetime

df['DEP_TIME'] = pd.to_datetime(df['DEP_TIME'], format='%H%M', errors='coerce').dt.time
df['ARR_TIME'] = pd.to_datetime(df['ARR_TIME'], format='%H%M', errors='coerce').dt.time

df[['DEP_TIME','ARR_TIME']]
```
